```python
player_info = pd.read_csv('playersdata_for_salaries.csv')
player_info.head()
```

```python
unique_ids = np.unique(player_info['_id'])
len(unique_ids)
```

```python
player_info = player_info[['_id', 'name']]
player_info.head()
```

Match names to players to annual salaries by player id

```python
df_player_salaries = player_salaries.merge(player_info, left_on='player_id', right_on='_id')
df_player_salaries.pop('_id')
df_player_salaries.pop('league')
```

```python
df_player_salaries.head()
```

```python
df_player_salaries.shape
```

```python
df_player_salaries.query("name == 'Aaron Brooks' & season_end == 2014")
```

Some players in the dataset have two different teams and salaries listed for a single season. This likely means that the player was traded mid-season. Since we do not know if the listed salaries are what the players were supposed to make or reflect salaries that were pro-rated based on the player's time spent playing for that team, we make the modeling assumption to **average** the salaries listed for a single player during a single season.

```python
df_player_salaries_averaged = df_player_salaries.groupby(["name", "season", "season_end", "season_start"])["salary"].mean()
df_player_salaries_averaged = df_player_salaries_averaged.reset_index()
```

```python
df_player_salaries_averaged.shape
```

Final name of player salary dataframe

```python
from google.colab import files
df_player_salaries_averaged.to_csv('salaries_by_season_new.csv') 
#files.download("salaries_by_season_new.csv")
```

```python
from datetime import datetime
date = '2006/12/29'
date=datetime.strptime(date, '%Y/%m/%d')
print(type(date))
if datetime.strptime('2005/10/25', '%Y/%m/%d') < date:
  print("True")
```

```python
def season_start_end(REAL_DATE):
  try:
    datetime.strptime(REAL_DATE, '%Y/%m/%d')
  except:
    REAL_DATE = REAL_DATE[0:4]+'/'+REAL_DATE[5:7]+'/'+REAL_DATE[8:]
  if datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2004/06/15', '%Y/%m/%d'):
    return 2003
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2004/11/02', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2005/06/23', '%Y/%m/%d'): #
    return 2004
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2005/11/01', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2006/06/20', '%Y/%m/%d'): # 
    return 2005
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2006/10/31', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2007/06/14', '%Y/%m/%d'): #
    return 2006
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2007/10/30', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2008/06/17', '%Y/%m/%d'): #
    return 2007
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2008/10/28', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2009/06/14', '%Y/%m/%d'): # 
    return 2008
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2009/10/27', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2010/06/17', '%Y/%m/%d'): #
    return 2009
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2010/10/26', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2011/06/12', '%Y/%m/%d'): #
    return 2010
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2011/12/25', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2012/06/21', '%Y/%m/%d'): #
    return 2011
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2012/10/30', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2013/06/20', '%Y/%m/%d'): #
    return 2012
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2013/10/29', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2014/06/15', '%Y/%m/%d'): #
    return 2013
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2014/10/28', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2015/06/16', '%Y/%m/%d'): #
    return 2014
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2015/10/27', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2016/06/19', '%Y/%m/%d'): #
    return 2015
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2016/10/25', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2017/06/12', '%Y/%m/%d'): #
    return 2016
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2017/10/17', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2018/06/08', '%Y/%m/%d'): #
    return 2017
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2018/10/16', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2019/06/13', '%Y/%m/%d'): #
    return 2018
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2019/10/22', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2020/10/11', '%Y/%m/%d'): #
    return 2019
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2020/12/22', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2021/07/20', '%Y/%m/%d'): #
    return 2020
  elif datetime.strptime(REAL_DATE, '%Y/%m/%d')>=datetime.strptime('2021/10/19', '%Y/%m/%d') and datetime.strptime(REAL_DATE, '%Y/%m/%d')<=datetime.strptime('2022/06/16', '%Y/%m/%d'): #
    return 2021

game_details = pd.read_csv('games_details.csv')
games = pd.read_csv('games.csv')
games.sample(10)

comp_game = pd.merge(game_details, games, how="left", on=['GAME_ID', 'GAME_ID']) #Taking games.csv and game_details.csv and combining them so that we can separate by date

comp_game[['GAME_YEAR', 'GAME_MONTH', 'GAME_DAY']] = comp_game['GAME_DATE_EST'].str.split('-', expand=True)

comp_game['REAL_GAME_YEAR'] = comp_game.apply(lambda x: season_start_end(x['GAME_DATE_EST']), axis=1)

comp_game.sample(10)
```
